```python
import pandas as pd
```

```python
##1. O seguinte nootbook tem a intençaõ de apresentar conceitos basicos estatisticos, para isso sera o datasest sobre obesidade.
##Sera aboradado variaveis qualitativas ao longo da demonstração.
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset/obesidade/ObesityDataSet_raw_and_data_sinthetic.csv')
```

```python
##1.1 Frequencia absoluta: A frequencia absoluta mede o quantos de vezes aparece um valor dentro de uma amosta. Para calcular ela basta
##    escolher qual variavel deseja contar o numero de ocorrencia.
df_frequecia_abs = df.groupby(df['NObeyesdad'])['NObeyesdad'].count()
df_frequecia_abs
```

```python
##1.2 Frequencia Relativa: A frequencia relativa é a porcentagem entre o total da amostra e o numero de vezes que determinado valor aparece.
##    Para calcular basta dividir a frequencia absoluta pelo o total.
df_frequecia_relativa = df['NObeyesdad'].value_counts(normalize=True)*100
```

```python
##1.3 Frequencia Relativa Acumulada: ela mostra como a frequência relativa de uma categoria se acumula a medida que percorre as categorias
##    da amostra.
df_frequecia_rela_acum = df_frequecia_relativa.cumsum()
df_frequecia_rela_acum
```

```python
tabela_frequencia = pd.concat([df_frequecia_abs, df_frequecia_relativa, df_frequecia_rela_acum], axis=1) #criando a tabela
tabela_frequencia.columns = ['Frequência Absoluta', 'Frequência Relativa (%)', 'Frequência Relativa Acumulada (%)'] #nomendo as colunas
tabela_frequencia = tabela_frequencia.sort_values(by='Frequência Relativa Acumulada (%)') #ordenando para ficar mais legivel
```
